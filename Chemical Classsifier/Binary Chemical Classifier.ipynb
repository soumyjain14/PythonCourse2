```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X_train=pd.read_csv("Training Data/Logistic_X_Train.csv")
```

```python
y_train=pd.read_csv("Training Data/Logistic_Y_Train.csv")
```

```python
X_test=pd.read_csv("Test Cases/Logistic_X_Test.csv")
```

```python
plt.style.use("seaborn")
plt.scatter(X_train.iloc[:1500,0],X_train.iloc[:1500,1],c='yellow',cmap=plt.cm.Accent)
plt.scatter(X_train.iloc[1500:,0],X_train.iloc[1500:,1],c='blue',cmap=plt.cm.Accent)
plt.scatter(X_train.iloc[:1500,2],X_train.iloc[:1500,2],c='red')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```
