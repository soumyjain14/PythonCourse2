```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=np.random.randn(1000)
```

```python
Y=np.random.randn(1000)
```

```python
#Doing reverse process
# generated random numbers with Mean 0 and STD 1, now creating other
# Usually datasets are like x1,y1 points
Mean=10
Std=2
X1=(X+Mean)*Std
```

```python
Y1=(Y+Mean)*Std
```

```python
plt.scatter(X1,Y1,c="yellow")
plt.scatter(X,Y,c="red")
plt.show()

# Yellow is not standard normal distribution
# Red is standard normal distribution, Mean=0, STD=1
```
