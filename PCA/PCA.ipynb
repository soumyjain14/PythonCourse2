```python
# PCA Algorithm
#Suppose there are m examples and n features so total matrix is of
# (m,n) size, and we have to reduce the size of features from n to k
# There are 4 steps involved in PCA algorithm
# 1) Preprocessing-Standardization i.e Mean=0 and SD=1
# 2) Compute the Covariance matrix (size = n*n)
# 3) Calculate the SVD- Singular Value Decomposition-for eigen vectors
# U(eigen vector)(all)= n*n (calculated for features)
# U reduced= U[:,:k], (n,k)
# 4) Projection, Z=U reduced *Transpose (dot product) X (feature)
```

```python
# X_train has too much data, so using X_test as X and reshaping it to flatten it
X=X_test.reshape(-1,28*28)
Y=y_test
```

```python
X.shape,Y.shape
```

```python
# Step1-Preprocessing-Using the inbuilt function of StandardScaler
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_=sc.fit_transform(X)
```

```python
plt.imshow(X_[90].reshape(28,28),cmap='gray')
plt.show()
```

### Sklearn PCA

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2) # reducing 784 dimensions to 2
```

```python
Z_pca=pca.fit_transform(X_)
```

```python
Z_pca.shape
```

```python
pca.explained_variance_ # If total was 100% var, this 2 combined make upto 70%of data
```

```python
# this was through inbuilt PCA, now using Custom PCA
```

### Custom PCA

```python
# Step 2, Covariance matrix, using trick of dot product of X_.T and X_
import numpy as np
```

```python
covar=np.dot(X_.T,X_)
```

```python
covar.shape # n dimension so covar of shape n,n
```

```python
## Step 3, Using the SVD, using inbuilt function of SVD
from numpy.linalg import svd
```

```python
# SVD return 3 things, U, S and V, but we require only U
```

```python
U,S,V=svd(covar)
```

```python
U.shape
```

```python
# now we need only 2 dimensions so,
U_reduced=U[:,:2]
```

```python
U_reduced.shape
```

```python
X_.shape
```

```python
# For 4th Step, to calculate Z, i.e projection
```

```python
Z=np.dot(X_,U_reduced)
```

```python
Z.shape
```
